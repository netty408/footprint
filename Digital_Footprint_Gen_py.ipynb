```
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
from matplotlib.gridspec import GridSpec

# ตั้งค่าฟอนต์ (พยายามหาฟอนต์ที่รองรับภาษาไทย หรือใช้ Default หากไม่พบ)
# ใน Kali Linux อาจจะต้องลงฟอนต์ไทยเพิ่ม (sudo apt install fonts-noto-thai)
# หากไม่พบฟอนต์ไทย กราฟจะแสดงผลเป็นสี่เหลี่ยม ให้เปลี่ยนข้อความในตัวแปรเป็นภาษาอังกฤษแทนได้
plt.rcParams['font.family'] = 'sans-serif'
# รายชื่อฟอนต์ไทยที่พบบ่อยใน OS ต่างๆ
font_list = ['Noto Sans Thai', 'Tahoma', 'Leelawadee', 'Garuda', 'Waree']
for font in font_list:
    if font in [f.name for f in fm.fontManager.ttflist]:
        plt.rcParams['font.family'] = font
        break

# การตั้งค่า Style แบบ Dark Cyber
plt.style.use('dark_background')
colors_cyber = ['#06b6d4', '#ec4899', '#8b5cf6', '#f97316'] # Cyan, Pink, Purple, Orange

# สร้างหน้ากระดาน (Figure)
fig = plt.figure(figsize=(16, 9))
fig.suptitle('Digital Persona Report: Jitkanya Layod', fontsize=24, color='#06b6d4', fontweight='bold')
gs = GridSpec(2, 3, figure=fig)

# --- ส่วนที่ 1: Pie Chart (Category Breakdown) ---
ax1 = fig.add_subplot(gs[0, 0])
labels_cat = ['Dev/Code', 'Music', 'Social', 'Commerce']
sizes_cat = [6, 4, 6, 2]
explode = (0.1, 0, 0, 0)

wedges, texts, autotexts = ax1.pie(sizes_cat, explode=explode, labels=labels_cat, autopct='%1.1f%%',
                                   shadow=True, startangle=90, colors=colors_cyber)
ax1.set_title('Account Distribution (การกระจายตัว)', fontsize=14, color='white')
plt.setp(texts, color="white")
plt.setp(autotexts, size=10, weight="bold")

# --- ส่วนที่ 2: Bar Chart (Platform Types) ---
ax2 = fig.add_subplot(gs[0, 1:])
platforms = ['Coding', 'Music Streaming', 'Global Market', 'Security/Hacking']
counts = [5, 4, 2, 1]
bars = ax2.bar(platforms, counts, color=colors_cyber)

ax2.set_title('Platform Risk Analysis (วิเคราะห์ความเสี่ยงแพลตฟอร์ม)', fontsize=14, color='white')
ax2.bar_label(bars, padding=3, color='white')
ax2.set_facecolor('#1e293b') # Dark card background

# --- ส่วนที่ 3: Radar Chart (Persona Mismatch) ---
# Radar chart ต้องใช้ Polar axes
ax3 = fig.add_subplot(gs[1, 0], polar=True)
categories = ['Content', 'Social', 'Dev Skills', 'CyberSec', 'Commerce']
categories = [*categories, categories[0]] # ปิดวงกลม

# ข้อมูลที่คาดหวัง (Influencer)
expected = [90, 85, 10, 5, 40]
expected = [*expected, expected[0]]

# ข้อมูลจริง (Actual)
actual = [50, 40, 80, 75, 50]
actual = [*actual, actual[0]]

label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(expected))

ax3.plot(label_loc, expected, label='Influencer Persona', color='#ec4899', linewidth=2, linestyle='--')
ax3.fill(label_loc, expected, color='#ec4899', alpha=0.2)

ax3.plot(label_loc, actual, label='Actual Footprint', color='#06b6d4', linewidth=2)
ax3.fill(label_loc, actual, color='#06b6d4', alpha=0.2)

ax3.set_title('Identity Mismatch (ความขัดแย้งของตัวตน)', fontsize=14, color='white', pad=20)
ax3.set_xticks(label_loc[:-1])
ax3.set_xticklabels(categories[:-1], color='white')
ax3.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1), fontsize='small')

# --- ส่วนที่ 4: Risk Meter (Text Summary) ---
ax4 = fig.add_subplot(gs[1, 1:])
ax4.axis('off')

summary_text = (
    "SUBJECT ANALYSIS SUMMARY:\n\n"
    "1. HIGH TECHNICAL CAPABILITY: พบรอยเท้าดิจิทัลในกลุ่ม Developer และ Security (HackenProof, GNOME)\n"
    "   ซึ่งขัดแย้งกับภาพลักษณ์ Influencer ท่องเที่ยวทั่วไป\n\n"
    "2. GEOGRAPHIC ANOMALY: มีการใช้งาน E-commerce ในโซนยุโรปและอเมริกาใต้ (Dealabs, Mercadolivre)\n\n"
    "3. RISK ASSESSMENT: MODERATE TO HIGH (65%)\n"
    "   ความเสี่ยง: อาจเป็นผู้เชี่ยวชาญที่ใช้อัตลักษณ์อื่นบังหน้า หรือบัญชีถูกนำไปใช้ทดสอบระบบ"
)

ax4.text(0.5, 0.5, summary_text, ha='center', va='center', fontsize=12, color='#e2e8f0',
         bbox=dict(boxstyle="round,pad=1", fc="#1e293b", ec="#06b6d4", alpha=0.8))

# Save file
plt.tight_layout()
output_filename = 'jitkanya_layod_analysis.png'
plt.savefig(output_filename, dpi=300, facecolor='#0f172a')
print(f"Exported analysis to: {output_filename}")

plt.show()
```
